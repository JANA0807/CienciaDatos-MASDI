```python
import glob
import pandas as pd
import numpy as np
import calendar
from google.colab import files
uploaded = files.upload()#Carga de archivos
```

```python
#Archivos a dt
dt_Prestamos=pd.read_csv('PRESTAMOS_1.F250312.CSV',encoding="latin1", sep=';') # Changed the delimiter to ';'. Remove or adjust error_bad_lines if needed.
dt_FNG=pd.read_excel('FNG.xlsx')
```

```python
Prestamos1= dt_Prestamos.drop(0)#se borra fila 1
Prestamos1['CCC '] = Prestamos1['CCC '].str.replace('"00130', '')
Prestamos1['CCC '] = Prestamos1['CCC '] .astype(str)
saldos=dt_FNG
saldos=saldos.rename(columns={'Número de pagaré':'CCC '})
saldos['CCC '] = saldos['CCC '].astype(str)
Prestamos1['Saldo_Total']=''
Prestamos1=Prestamos1[['VR_INICIAL ','CALIFICACION_DEF ','SALDO_K ','Saldo_Total','SALDO_I ', 'CCC ','FECHA_VTO ','NUMERO ']]
Prestamos1['SALDO_K '] = Prestamos1['SALDO_K '].str.replace(' ', '').str.replace(',', '').astype(float)
Prestamos1['SALDO_I '] = Prestamos1['SALDO_I '].str.replace(' ', '').str.replace(',', '').astype(float)
Prestamos1['Saldo_Total'] = Prestamos1['SALDO_K '] + Prestamos1['SALDO_I ']
Prestamos1['FECHA_VTO '] = pd.to_datetime(Prestamos1['FECHA_VTO '], errors='coerce')
Prestamos1['FECHA_VTO ']=Prestamos1['FECHA_VTO '].dt.strftime("%d/%m/%Y")
Prestamos1
```

```python
Prestamos1['CCC '] = Prestamos1['CCC '].astype(str)
Prestamos1['CCC '] = Prestamos1['CCC '].str.strip()
Prestamos1['CCC '] = Prestamos1['CCC '].str.zfill(13)

saldos['CCC '] = pd.to_numeric(saldos['CCC '], errors='coerce').fillna(0).astype(int).astype(str)
saldos['CCC '] = saldos['CCC '].str.strip()
saldos['CCC '] = saldos['CCC '].str.zfill(13)

cruce1=pd.merge(saldos,Prestamos1,on='CCC ', how='inner')
# Reemplazar la coma por un punto para que se reconozca como número decimal
cruce1['VR_INICIAL '] = cruce1['VR_INICIAL '].str.replace(',', '').astype(float)
# Redondear la columna 'VR_INICIAL' a 0 decimales
#cruce1['VR_INICIAL '] = cruce1['VR_INICIAL '].round(0)
cruce1['VR_INICIAL '] = np.ceil(cruce1['VR_INICIAL ']).astype(int)
cruce1.head()
```

```python
cruce1['Calificación de riesgo de la obligación']=cruce1['CALIFICACION_DEF ']
cruce1['Saldo de capital'] = cruce1['SALDO_K ']
cruce1['Saldo total']= cruce1['Saldo_Total']
cruce1['Fecha de corte']=pd.Timestamp("28/02/2025")
cruce1['Fecha de corte'] = cruce1['Fecha de corte'].dt.strftime("%d/%m/%Y")
cruce1['Fecha de corte'] = pd.to_datetime(cruce1['Fecha de corte'], errors='coerce', format='%d/%m/%Y')

# Calcular la diferencia de días y dividir por 30
# Obtener la fecha actual como Timestamp
cruce1['FECHA_VTO '] = pd.to_datetime(cruce1['FECHA_VTO '], errors='coerce', format='%d/%m/%Y')
cruce1['FechadeHoy']=pd.to_datetime("13/03/2025", format='%d/%m/%Y')
cruce1['FechadeHoy'] = pd.to_datetime(cruce1['FechadeHoy'])
cruce1['FechadeHoy'] = pd.to_datetime(cruce1['FechadeHoy'],errors= 'coerce', format='%d/%m/%Y')
cruce1['diferencia_dias'] = (cruce1['FechadeHoy'] - cruce1['FECHA_VTO ']).dt.days/30
cruce1['Número cuotas mora'] = np.where(cruce1['Saldo total'] > 0.000001,np.where(cruce1['FECHA_VTO '] < cruce1['FechadeHoy'],cruce1['diferencia_dias'],0),'')
cruce1['Fecha inicio mora']= cruce1['FECHA_VTO ']
cruce1['Fecha Cancelación'] = np.where(cruce1['Saldo total'] == 0,np.where(cruce1['FECHA_VTO '] > cruce1['FechadeHoy'],cruce1['FechadeHoy'].dt.strftime('%d/%m/%Y'),cruce1['FECHA_VTO '].dt.strftime('%d/%m/%Y')), pd.NaT)
#----------------------------------------------------------------------------------------------------------------------------------------
cruce1['Número cuotas mora'] = np.where(cruce1['Saldo de capital'] == 0,'',cruce1['Número cuotas mora'])
cruce1['Fecha inicio mora'] = np.where(cruce1['Saldo de capital'] == 0,pd.NaT,cruce1['Fecha inicio mora'].dt.strftime('%d/%m/%Y'))
mes_referencia = cruce1['Fecha de corte'].dt.month[0]
year_referencia=cruce1['Fecha de corte'].dt.year[2]
cruce1['Fecha vencimiento prestamo'] = pd.to_datetime(cruce1['Fecha vencimiento prestamo'],errors= 'coerce', format="%d/%m/%Y")
cruce1['Fecha Cancelación'] = pd.to_datetime(cruce1['Fecha Cancelación'], format="%d/%m/%Y")
cruce1['Fecha Cancelación'] = np.where((cruce1['Saldo de capital'] == 0)&(cruce1['Fecha Cancelación'].dt.month!=mes_referencia) | (cruce1['Fecha Cancelación'].dt.year!=year_referencia),cruce1['Fecha de corte'],cruce1['Fecha Cancelación'])

#cruce1['Fecha Cancelación'] = pd.to_datetime(cruce1['Fecha Cancelación'], unit='ns').dt.strftime('%d/%m/%Y')
cruce1['Fecha Cancelación'] = np.where((cruce1['Saldo de capital'] != 0),pd.NaT,cruce1['Fecha Cancelación'].dt.strftime('%d/%m/%Y'))
cruce1=cruce1.rename(columns={'Unnamed: 31':'Calculo V'})
cruce1['Calculo V']=np.where((cruce1['Saldo de capital'] != 0),(cruce1['Valor desembolso'] - cruce1['Saldo de capital']),cruce1['Calculo V'])

cruce1['Saldo de capital']=np.where((cruce1['Saldo de capital'] != 0) & ((cruce1['Calculo V'] > -10) & (cruce1['Calculo V'] < 10)),cruce1['Valor desembolso'],cruce1['Saldo de capital'])
cruce1['Calculo V']=np.where((cruce1['Saldo de capital'] != 0),(cruce1['Valor desembolso'] - cruce1['Saldo de capital']),cruce1['Calculo V'])
cruce1['OBSERVACIONES ']=np.where((cruce1['Saldo de capital'] != 0) & (cruce1['Calculo V']!=0)&(cruce1['Valor desembolso']== cruce1['VR_INICIAL '])&(cruce1['Saldo de capital']==cruce1['SALDO_K '])&(cruce1['Saldo total']==cruce1['Saldo_Total']),'OPERACION CON SALDO CORRECTO',cruce1['OBSERVACIONES '])
cruce1['OBSERVACIONES ']=np.where((cruce1['Saldo de capital'] != 0) &(cruce1['Calculo V']!=0)&(cruce1['Valor desembolso']!= cruce1['VR_INICIAL ']),'Revisar FNG en correo',cruce1['OBSERVACIONES '])
cruce1['Número cuotas mora'] = pd.to_numeric(cruce1['Número cuotas mora'], errors='coerce')
cruce1['OBSERVACIONES '] = np.where((cruce1['Número cuotas mora'].round(0)!= 0 ) & (cruce1['Número cuotas mora'].notna())&(cruce1['Saldo de capital'] != 0)&((cruce1['Fecha vencimiento prestamo']-cruce1['FECHA_VTO ']).dt.days < 10),'OPERACION CON SALDO Y CON MORA',cruce1['OBSERVACIONES '])
cruce1['OBSERVACIONES '] = np.where((cruce1['Número cuotas mora'].round(0)!= 0 ) & (cruce1['Número cuotas mora'].notna())&(cruce1['Saldo de capital'] != 0)&((cruce1['Fecha vencimiento prestamo']-cruce1['FECHA_VTO ']).dt.days > 10),'REVISAR FECHA DE VENCIMIENTO',cruce1['OBSERVACIONES '])

cruce1['OBSERVACIONES '] = np.where((cruce1['Saldo de capital'] != 0) & # Condición principal: Saldo de capital diferente de 0
    (cruce1['Fecha vencimiento prestamo'].dt.month <= mes_referencia) &  # Condición 1: Mes anterior al de referencia
    (cruce1['Fecha vencimiento prestamo'].dt.year <= year_referencia),# Condición 2: Año menor o igual al de referencia
    np.where(
        (abs((cruce1['Fecha vencimiento prestamo']-cruce1['FECHA_VTO ']).dt.days) > 10),  # Condición 3: Diferencia mayor a 10 días
        'REVISAR FECHA DE VENCIMIENTO',  # Resultado si se cumple la condicion
        'OPERACION CON SALDO Y CON MORA'   # Resultado si no se cumplen la condicion pero sí las 3 principales
    ),
    cruce1['OBSERVACIONES ']  # Valor si no se cumple la condición principal (Saldo de capital = 0)
)
cruce1['OBSERVACIONES '] = np.where(cruce1['Saldo de capital'] == 0 | (cruce1['Saldo de capital'].isnull()),'OPERACION CON SALDO 0.00',cruce1['OBSERVACIONES '])
cruce1['OBSERVACIONES ']
cruce1[['Número cuotas mora','Valor desembolso','VR_INICIAL ','Calculo V','Saldo de capital','SALDO_K ','Saldo total','Saldo_Total','OBSERVACIONES ']].head(492)
```
